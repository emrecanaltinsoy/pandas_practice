# Update rows and columns

### Change column names
- df.columns = [column_name_list] -> updates column names

- df.columns = [x.upper() for x in df.columns] -> update column names using list comprehension

- df.columns = df.columns.str.replace(" ", "_") -> replace spaces with underscores

- df.rename(columns={old_val1:new_val1, old_val2:new_val2...}, inplace=False) -> change specific column names using dictionary

### Change values in one index and multiple columns
- df.loc[index] = [list_of_values] -> change the values of a specific row (list should have values for each column)
- df.loc[index, [column1, column2]] = [list_of_values] -> change the values of a specific row and columns

### Change value at one specific location (row, column)
- df.loc[index, column1] = value -> change a specific row and a column
- df.at[index, column1] = value -> change a specific row and a column

### Change values in one whole column (apply, applymap, map, )
- df[column_name] = df[column_name].str.lower() -> lowercase all the values in a column

- df[column_name] = df[column_name].apply(lambda x: x.lower() -> lowercase all the values in a column

- df[column_name] = df[column_name].apply(func) -> apply func to a column

- df.apply(len, axis="rows") -> get the length of each column (number of rows)

- df.apply(len, axis="columns") -> get the length of each row (number of columns)

- df.applymap(func) -> apply func to every element in the dataframe

- df[column_name] = df[column_name].map(dict_of_old_new_values) -> replace values with new ones (returns NAN if item is not in the dict

- df[column_name] = df[column_name].replace(dict_of_old_new_values) -> replace values with new ones


### Concat

- new_df = pd.concat([df1, df2], axis="columns", sort=False) -> concat two dataframes

## Sorting Values

- df.sort_values(by=column_name, ascending=True, inplace=False) -> sort dataframe by given column name, (default) ascending=True, (default) inplace=False

- df.sort_values(by= [column1, column2, ...], ascending=True, inplace=False) -> sort dataframe by given column names, (default) ascending=True, (default) inplace=False

- df.sort_values(by= [column1, column2, ...], ascending=[bool1, bool2], inplace=False) -> sort dataframe by given column names, different sorting direction for each column, (default) inplace=False

- df.sort_index() -> sort dataframe by the index values

- df[column_name].sort_values() -> sort the values in a column

## Grouping and Aggregating

### median(), describe(), count(), value_counts()

- df[column].median() -> ignores NAN and returns median of column
- df[[column1, column2]].median() -> ignores NAN and returns median of multiple columns


- df.median() -> ignores NAN and returns median for all numerical columns


- df.describe() -> returns description of numerical columns


- df[column].count() -> ignores NAN and returns number of values
- df[column].value_counts() -> returns number of each item
- df[column].value_counts(normalize=True) -> returns percentage of each item

### Group by

- groups = df.groupby([column_name]) -> group by column values
- groups.get_group(column_value) -> get the group for a specific value

- groups[another_column_name].value_counts() -> get the value counts for a specific column in a group
- groups[another_column_name].median() -> get the median for a specific column in a group


### Agg

- groups[another_column_name].agg(["median", "mean"]) -> get median and mean for the group for a given another column name

## Cleaning Data

- df.dropna(axis="index", how="all") -> drops rows which have all NAN values
- df.dropna(axis="columns", how="any") -> drops columns which have any NAN values
- df.dropna(axis="columns", how="any") -> drops columns which have all NAN values

- df.dropna(axis="index", how="any", subset=[column1, column2]) -> checks subset columns for NAN values and drops rows

### replace values as NAN

- df.replace("NA", np.nan, inplace=True) -> replace any "NA" value as NAN
- df.replace("Missing", np.nan, inplace=True) -> replace any "Missing" value as NAN
